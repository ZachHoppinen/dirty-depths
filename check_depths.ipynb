```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
import geopandas as gpd

df = gpd.read_file('snowex_depths.csv', GEOM_POSSIBLE_NAMES="geometry", KEEP_GEOM_COLUMNS="NO")

df.depth = df.depth.astype(float, errors = 'raise')
```

```python
df.depth.plot(kind = 'hist', bins = 150)
plt.xlabel('Depth')
```

```python
# import contextily as ctx
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
for i, loc in enumerate(df.Location.unique()):
    ax = axes.ravel()[i]
    df[df.Location == loc].plot(ax = ax)
    ax.set_title(loc)
    ax.ticklabel_format(useOffset=False)
plt.tight_layout()
```
